```python
def get_rsi_signal(rsi):
    if rsi > 80:
        return 'Sell'
    elif rsi < 20:
        return 'Buy'
    else: 
        return 'Hold'

df['RSI_Signal'] = df['RSI_21'].apply(get_rsi_signal)

df['RSI_Signal'].tail(30)

```

```python
df['Stop_loss'] = df['ATRr_10'] * 2

df['Macd_signal'] = np.where(df['MACD'] > df['MACD_S'], 'Buy',
                             np.where(df['MACD'] < df['MACD_S'], 'Sell', 'Hold'))

df
```

```python

df.drop(columns=df[['Signals_rsi', 'ATRr_20', 'Bbands_signal']])

df['Ema_signal'] = np.where(df['Close'] > df['EMA_50'], 'Buy',
                            np.where(df['Close'] < df['EMA_50'], 'Sell', 'Hold'))
df['Sma_signal'] = np.where(df['Close'] > df['SMA_50'], 'Buy',
                            np.where(df['Close'] < df['SMA_50'], 'Sell', 'Hold'))
# giving false signals for some reason
df['Bbands_signal'] = np.where(df['Close'] < df['BBL'], 'Buy',
                            np.where(df['Close'] > df['BBU'], 'Sell', 'Hold'))
# df['OBV_Signal'] = np.where(df['OBV'] > df['EMA_20'], 'Buy',
#                      np.where(df['OBV'] < df['EMA_20'], 'Sell', 'Hold'))

df.tail()
```

```python
# df.drop(columns=df[['Signals_rsi', 'ATRr_20', 'Bbands_signal']])
df['Signal'] = np.where(
    (df['Ema_signal'] == 'Buy') & (df['Macd_signal'] == 'Buy'), 'Buy',
    np.where((df['Ema_signal'] == 'Sell') & (df['Macd_signal'] == 'Sell'),
             'Sell', 'Hold')
)

print(np.where(df['Signal'] == 'Buy'))
```

```python
df['Signal'] = np.where((df['Signal'] == 'Buy'), 1,
                        np.where((df['Signal'] == 'Sell'), -1, 0))

print(np.where(df['Signal'] == -1))
```

```python

# Initialize columns
df['Position'] = 0.0
df['Cash'] = 0.0
df['Portfolio'] = 0.0
df['Stop_Loss_Hit'] = False  # Track stop loss exits

initial_cash = 100000
cash = initial_cash
position = 0.0
entry_price = 0.0


# Backtest loop
for i in range(1, len(df)):
    price = df['Close'].iloc[i]
    signal = df['Signal'].iloc[i]
    atr_stop = df['Stop_loss'].iloc[i]

    # BUY
    if signal == 1 and position == 0:
        position = cash / price
        cash = 0
        entry_price = price
        stop_price = entry_price - atr_stop

    # SELL (normal signal)
    elif signal == -1 and position > 0:
        cash = position * price
        position = 0
        entry_price = 0.0

    # STOP LOSS exit
    elif position > 0 and price < stop_price:
        cash = position * price
        position = 0
        entry_price = 0.0
        df.at[df.index[i], 'Stop_Loss_Hit'] = True

    df.at[df.index[i], 'Position'] = position
    df.at[df.index[i], 'Cash'] = cash
    df.at[df.index[i], 'Portfolio'] = cash + position * price

df['Portfolio'] = df['Portfolio'].ffill()
# df['Returns'] = np.log(df['Portfolio'] / df['Portfolio'].shift(1)).fillna(0)
returns = df['Portfolio'].pct_change()
returns = returns.replace([np.inf, -np.inf], np.nan).dropna()

if not returns.empty and returns.std() != 0 and not np.isnan(returns.std()):
    sharpe_ratio = (returns.mean() / returns.std()) * np.sqrt(252)
else:
    sharpe_ratio = 0




# Metrics
total_return = (df['Portfolio'].iloc[-1] - initial_cash) / initial_cash * 100
# sharpe_ratio = df['Returns'].mean() / df['Returns'].std()
max_drawdown = ((df['Portfolio'].cummax() - df['Portfolio']) / df['Portfolio'].cummax()).max() * 100
num_trades = df['Signal'].diff().abs().sum() / 2  # Approx number of buy/sell pairs

profitable_trades = []
in_position = False
buy_price = 0

for i in range(len(df)):
    sig = df['Signal'].iloc[i]
    price = df['Close'].iloc[i]
    if sig == 1 and not in_position:
        buy_price = price
        in_position = True
    elif sig == -1 and in_position:
        profitable_trades.append(price > buy_price)
        in_position = False

percent_profitable = (np.sum(profitable_trades) / len(profitable_trades) * 100) if profitable_trades else 0

# Print metrics
print(f"Total Return: {total_return:.2f}%")
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Max Drawdown: {max_drawdown:.2f}%")
print(f"Number of Trades: {num_trades}")
print(f"Percent Profitable: {percent_profitable:.2f}%")
```
